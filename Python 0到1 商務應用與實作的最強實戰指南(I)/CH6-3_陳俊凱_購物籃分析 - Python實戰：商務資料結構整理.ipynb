<span style="font-size:0.7cm; line-height: 1cm;font-family:cursive;"> 1.引入套件與資料</span><br>

```python
from apyori import apriori
import pandas as pd
import numpy as np
alldata = pd.read_csv('sales_data.csv')
alldata = pd.DataFrame(alldata)
```

<span style="font-size:0.7cm; line-height: 1cm;font-family:cursive;"> 2.篩選前三銷售量好的商品</span><br>

```python
top3_series = alldata.groupby('系列')['單價'].sum()#[0:3]
top3_series=top3_series.sort_values(ascending= False)
top3_series = top3_series.index[0:3]
```

<span style="font-size:0.7cm; line-height: 1cm;font-family:cursive;"> 3.利潤表格製作</span><br>

```python
profit_df=alldata.copy()
profit_df =profit_df.drop_duplicates('產品') 
profit_df =profit_df[['產品','單價','成本']] 
profit_df['利潤'] =profit_df['單價']-profit_df['成本']
```

<span style="font-size:0.7cm; line-height: 1cm;font-family:cursive;"> 4. 取出系列銷售資料-舉例

```python
# 取出包含系列銷售資料
# 系列1舉例

record=[] 
s='系列1'

series_data=alldata[alldata['系列']==s]
# 取出包含系列1的訂單編號

order_number=np.unique(series_data['訂單編號'])
```

```python
# 取出包含系列 1 第 1 個訂單編號內購買資料

i = order_number[1]
series_data[series_data['訂單編號']==i]
```

```python
# 取出包含系列 1 第 1 個訂單編號內產品項目

ex_cart=series_data[series_data['訂單編號']==i]['產品'].values
print(ex_cart)
```

<span style="font-size:0.7cm; line-height: 1cm;font-family:cursive;"> 5. 取出三個系列下所有購物籃資料

```python
# 取出三個系列下各訂單內產品項目

record=[] 

for s in top3_series:
    series_data=alldata[alldata['系列']==s]
    order_number=np.unique(series_data['訂單編號'])
    for i in order_number:
        cart =series_data[series_data['訂單編號']==i]['產品'].values
        record.append(cart)
        print(cart)
```
