## 一、基本語法
### 1.	關鍵字有三種：「if」、「elif」及「else」。
### 2.	可以有零個或多個 elif 部分，以及一個可寫可不寫的 else 部分。
### 3.	if-elif-else條件式(Condition)最後需以「：」結尾。
### 4.	程式執行(Statement)的部分要縮排。


```python
if condition1: # 條件一
    statement1
elif condition2: # 條件二
    statement2
elif condition3: # 條件三
    statement3
else: # 其他情形
    statement4
```

## 二、應用

```python
score = 70 # 設定成績變數

if score >= 60: # 60分以上及格
    print("及格！")
else: # 不符合以上條件的時候執行
    print("不及格！")
```

```python
# 加入elif敘述

score = 70 # 設定成績變數

if (score > 80):
    print("A")
elif (score >= 70) & (score < 80): # 加上括號分群幫助易讀
    print("B")
elif (score >= 60) & (score < 70):
    print("C")
else:
    print("F")
```
