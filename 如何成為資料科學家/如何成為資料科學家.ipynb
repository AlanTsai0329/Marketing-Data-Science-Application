```python
from wordcloud import (WordCloud, get_single_color_func)

#PS : Credits to Andreas Mueller, creator of wordcloud, for the following code of the class 'GroupedColorFunc'.
#He made the code fully public for people who want to use specific color for specific words and made an example.
#Source link : https://amueller.github.io/word_cloud/auto_examples/colored_by_group.html#sphx-glr-auto-examples-colored-by-group-py

class GroupedColorFunc(object):

    def __init__(self, color_to_words, default_color):
        self.color_func_to_words = [
            (get_single_color_func(color), set(words))
            for (color, words) in color_to_words.items()]

        self.default_color_func = get_single_color_func(default_color)

    def get_color_func(self, word):
        """Returns a single_color_func associated with the word"""
        try:
            color_func = next(
                color_func for (color_func, words) in self.color_func_to_words
                if word in words)
        except StopIteration:
            color_func = self.default_color_func

        return color_func

    def __call__(self, word, **kwargs):
        return self.get_color_func(word)(word, **kwargs)


#############################################################
# Get text data from the free
text = free[pd.notnull(free["KaggleMotivationFreeForm"])]["KaggleMotivationFreeForm"]

wc = WordCloud(collocations=False,height=800, width=600,  relative_scaling=0.2,random_state=74364).generate(" ".join(text))

color_to_words = {
    # words below will be colored with a green single color function
    '#00ff00': ['data science', 'mining', 'big',
                'bigdata', 'machine', 'learning']
}

# Words that are not in any of the color_to_words values will be colored with grey
default_color = 'grey'

# Create a color function with multiple tones
grouped_color_func = GroupedColorFunc(color_to_words, default_color)

# Apply our color function
wc.recolor(color_func=grouped_color_func)

# Plot
plt.figure(figsize=(12,12))
plt.imshow(wc, interpolation='bilinear')
plt.axis("off");
```

字體越大，越鮮明，代表越多人使用，讓我們看看他們之間的關聯 :

**前處理與編碼轉換**

由於kaggle的資料含有類別變數，而再進行機器學習時，我們必須要將類別變數轉換成可分析的數值變數，那以下有兩種熱門的方法


* **標籤編碼** : 該方法將各類別轉換成數字，在商業界及社會科學中，我們最常見的就是李克特5點量表的轉換，舉例來說：科技業就轉換成1；學術界就轉換成2，諸如此類，只不過這僅限「順序尺度」，也就是具有大小之分的類別，方能適用該法。


* **獨熱編碼** : 也就是我們俗稱的One Hot encoding，這代表將類別轉換成欄位格式，並且以1（有該類別）與0（沒有該類別）的形式表示之

Kaggle資料集則適用獨熱編碼，因此我們會以pd.get_dummies一行程式碼立即取得所要之編碼結果


```python
temp=data_dem

target = temp['CompensationAmount']
target.replace(to_replace=[1,2], value=[0,1],inplace=True )
temp.drop('CompensationAmount',axis=1,inplace=True)
temp2=pd.get_dummies(data=temp,columns=list(temp))
```

嘿嘿～ 我們就可以看到EmployerSize原本是有20到99、5000到9999等選項，現將其都轉換為「欄位」，並以1與0與的方式呈現之

**建模啦！**

為了從羅吉斯迴歸中找出真正重要的變數，所以建議在做羅吉斯迴歸時
* 可以使用L1正規化，來調整係數，這樣也方便我們知道越靠進0的係數，月不重要
* 調整C的大小可以來限制模型係數的程度，所以可以用小的C，讓資料變得很稀疏，方便我們知道重要因子

```python
from sklearn.metrics import f1_score, precision_score, recall_score , accuracy_score, roc_curve, auc, classification_report
from sklearn.linear_model import LogisticRegression, RidgeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier
from sklearn import cross_validation
from sklearn.model_selection import cross_val_score
import operator


clf = LogisticRegression(penalty='l1', C=0.05, tol=0.0001, random_state=42, solver='liblinear')
clf.fit(X_train,y_train)

coefs=np.transpose(clf.coef_)

coef_l1_LR = clf.coef_.ravel()
sparsity_l1_LR = np.mean(coef_l1_LR == 0) * 100
print("Sparsity achieved: %.2f%%" % sparsity_l1_LR)

feature_importance={}
for (feature,k) in zip(list(X_train),range(len(coefs))) : 
        feature_importance[feature]=abs(coefs[k])
sorted_features = sorted(feature_importance.items(), key=operator.itemgetter(1))
top10 = sorted_features[-10::]
top10
```

從top 10中重要的屬性中，可以發現
* **年資**
* **年齡**
* **資料分析師**

```python
import itertools
plt.subplots(figsize=(22,10))
time_spent=['TimeFindingInsights','TimeVisualizing','TimeGatheringData','TimeModelBuilding']
length=len(time_spent)
for i,j in itertools.zip_longest(time_spent,range(length)):
    plt.subplot((length/2),2,j+1)
    plt.subplots_adjust(wspace=0.2,hspace=0.5)
    scientist[i].hist(bins=10,edgecolor='black')
    plt.axvline(scientist[i].mean(),linestyle='dashed',color='r')
    plt.title(i,size=20)
    plt.xlabel('% Time')
plt.show()
```

<a id="point"></a>

1.** 學Python ** Python 與 R 已經雌雄爭霸近10年，在關鍵的2017年的各大調查中，Python終於勝出了R，且我們在第二單元可以看到Python實則已經打敗R，就算是商業領域，Python與R也早已經不相上下，未來的趨勢將會一面倒向Python，所以要入門的讀者，請選擇Python吧。

2.** 可以主修資料科學領域 ** 除了大部分主修數學或者電腦科學的人在資料科學都倍受青睞，這群人也了解演算法背後的知識。

3.** 學做專案、線上課程、打Kaggle ** 就像我們在第2、3章看到的，Kaggle、線上課程與專案都是可以讓新進的讀者們快速上手的資料科學專案，讓自己快速成長為資料科學家

4.** 學常用的機器學習演算法 ** 越常用的演算法也就代表他的表現效能越佳或易學，尤其是2、3章分析出來的羅吉斯迴歸、樹狀模型(隨機森林、決策樹等)、。

5.** 深度學習 ** 是未來發展AI非常重要的基石，讀者有了機器學習的知識後，就可以更深入得來探討深度學習領域，對未來工作領域會非常有幫助，建議可以從keras, Pytorch, tensorflow開始學起
