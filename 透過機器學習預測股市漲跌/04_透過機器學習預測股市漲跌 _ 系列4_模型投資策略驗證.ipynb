```python
import pandas as pd
import random
from datetime import datetime
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('回測.csv', usecols = ['Date', 'Label', 'fluctuation','Prediction'])
df['Date'] = pd.to_datetime(df['Date'])
df
```

```python
# 模型投資策略:遵從模型建議投資之逐月報酬率
AP = 1
month_r = int(2)
percentage_r = []
for i in range(0,378):
    if df.iloc[i,1] == df.iloc[i, 2]:
        AP = AP * (1 + abs(df.iloc[i,-1]))
    else:
        AP = AP * (1 - abs(df.iloc[i,-1]))
    if month_r == 13: #驗證時間橫跨兩年，故重置month
        month_r = 1
    if df['Date'][i].month == month_r:
        percentage_r.append((AP)-1)
        month_r += 1
        

percentage_r #後續將報酬率資料在excel中進行視覺化處理
```

```python
# 隨機投資策略:隨機分配各一半天數為看空(0)及看多(1)
randoms = []
for i in range(189):
    randoms.append(0)
for j in range(189):
    randoms.append(1)
random.shuffle(randoms)
df['Randoms'] = randoms
df
```

```python
#隨機投資策略:計算逐月報酬率
num = 1
percentage = []
month = int(2)
for i in range(len(df)):
    if df['Randoms'][i] == 0: #隨機猜測做空
        if df['fluctuation'][i] > 0:
            num = num * (1 - (df['fluctuation'][i]/100))
        else:
            num = num * (1 - (df['fluctuation'][i]/100))
    if df['Randoms'][i] == 1: #隨機猜測做多
        if df['fluctuation'][i] > 0:
            num = num * (1 + (df['fluctuation'][i]/100))
        else:
            num = num * (1 + (df['fluctuation'][i]/100))
    if month == 13: #驗證時間橫跨兩年，故重置month
        month = 1
    if df['Date'][i].month == month:
        percentage.append(num)
        month += 1
        
for i in range(len(percentage)):
    percentage[i] = percentage[i].round(4)
    percentage[i] = (percentage[i] - 1)*100

percentage #後續將報酬率資料在excel中進行視覺化處理


```
